```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

import scipy.signal as sp

import soundfile as sf
from scipy.signal import butter, filtfilt
import librosa
import librosa.display

# PARMETROS
sample_rate = 44100 # Sample rate in Hz
duration = 5 # Total duration in seconds
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
tempo_ajustado = np.linspace(0, duration, int(8*sample_rate * duration), endpoint=False)
silencio = 0

# CRIAO DO RUDO
amplitude_ruido = 0.08  # Amplitude do rudo
ruido = amplitude_ruido * np.random.randn(int(1 * (len(t) / 20)))

# CRIAO DE ONDAS SENOIDAIS
silencioNota  = np.sin(2 * np.pi *             silencio           *                 t[:int(1 * (len(t) / 20))])
#----------------------------------------------------------------------------------------------------------------------
C5_Onda        = np.sin(2 * np.pi *             C5                 *                 t[:int(1 * (len(t) / 20))])
C3_Onda        = np.sin(2 * np.pi *             C3                 *                 t[:int(1 * (len(t) / 20))])
G5_Onda        = np.sin(2 * np.pi *             G5                 *                 t[:int(1 * (len(t) / 20))])
B6_Onda        = np.sin(2 * np.pi *             B6                 *                 t[:int(1 * (len(t) / 20))])
Fsus5_Onda     = np.sin(2 * np.pi *             Fsus5              *                 t[:int(1 * (len(t) / 20))])
D3_Onda        = np.sin(2 * np.pi *             D3                 *                 t[:int(1 * (len(t) / 20))])
Fsus3_Onda     = np.sin(2 * np.pi *             Fsus3              *                 t[:int(1 * (len(t) / 20))])
B3_Onda        = np.sin(2 * np.pi *             B3                 *                 t[:int(1 * (len(t) / 20))])
Csus3_Onda     = np.sin(2 * np.pi *             Csus3              *                 t[:int(1 * (len(t) / 20))])
Csus4_Onda     = np.sin(2 * np.pi *             Csus4              *                 t[:int(1 * (len(t) / 20))])
E3_Onda        = np.sin(2 * np.pi *             E3                 *                 t[:int(1 * (len(t) / 20))])
B5_Onda        = np.sin(2 * np.pi *             B5                 *                 t[:int(1 * (len(t) / 20))])
G3_Onda        = np.sin(2 * np.pi *             G3                 *                 t[:int(1 * (len(t) / 20))])
#----------------------------------------------------------------------------------------------------------------------
G3_E3_Onda     = np.sin(2 * np.pi *             G3                 *                 t[:int(1 * (len(t) / 20))]) + np.sin(2 * np.pi *             E3                 *                 t[:int(1 * (len(t) / 20))])
E3_D3_Onda     = np.sin(2 * np.pi *             E3                 *                 t[:int(1 * (len(t) / 20))]) + np.sin(2 * np.pi *             D3                 *                 t[:int(1 * (len(t) / 20))])
E3_Fsus3_Onda  = np.sin(2 * np.pi *             E3                 *                 t[:int(1 * (len(t) / 20))]) + np.sin(2 * np.pi *             Fsus3              *                 t[:int(1 * (len(t) / 20))])
C3_G3_Onda     = np.sin(2 * np.pi *             C3                 *                 t[:int(1 * (len(t) / 20))]) + np.sin(2 * np.pi *             G3                 *                 t[:int(1 * (len(t) / 20))])
C3_B4_Onda     = np.sin(2 * np.pi *             C3                 *                 t[:int(1 * (len(t) / 20))]) + np.sin(2 * np.pi *             B4                 *                 t[:int(1 * (len(t) / 20))])
#----------------------------------------------------------------------------------------------------------------------


#--------------APLICAR RUDO NAS ONDAS-------------

C5_OndaR = C5_Onda + ruido; C3_OndaR = C3_Onda + ruido; G5_OndaR = G5_Onda + ruido; B6_OndaR = B6_Onda + ruido; Fsus5_OndaR = Fsus5_Onda + ruido;
G3_E3_OndaR = G3_E3_Onda + ruido; G3_OndaR = G3_Onda + ruido; E3_D3_OndaR = E3_D3_Onda + ruido; E3_Fsus3_OndaR = E3_Fsus3_Onda + ruido; D3_OndaR = D3_Onda + ruido;
C3_G3_OndaR = C3_G3_Onda + ruido; C3_B4_OndaR = C3_B4_Onda + ruido; Fsus3_OndaR = Fsus3_Onda + ruido; B3_OndaR = B3_Onda + ruido; Csus3_OndaR = Csus3_Onda + ruido;
Csus4_OndaR = Csus4_Onda + ruido; Csus3_OndaR = Csus3_Onda + ruido; E3_OndaR = E3_Onda + ruido; B5_OndaR = B5_Onda + ruido

#--------------------------------------------------
Compasso0 = np.concatenate((C5_OndaR,
silencioNota, G5_OndaR, silencioNota, B6_OndaR, silencioNota, Fsus5_OndaR,silencioNota,
C5_OndaR, silencioNota, G5_OndaR, silencioNota, B6_OndaR, silencioNota, Fsus5_OndaR,silencioNota,
C5_OndaR, silencioNota, G5_OndaR, silencioNota, B6_OndaR, silencioNota, Fsus5_OndaR,silencioNota,
C5_OndaR, silencioNota, G5_OndaR, silencioNota, B6_OndaR, silencioNota, Fsus5_OndaR,silencioNota))

Compasso1 = np.concatenate((np.concatenate((G3_E3_OndaR,G3_E3_OndaR,G3_E3_OndaR,G3_E3_OndaR)) + np.concatenate((C5_OndaR,silencioNota, G5_OndaR,silencioNota)),
                            np.concatenate((G3_E3_OndaR,G3_E3_OndaR,G3_E3_OndaR,G3_E3_OndaR)) + np.concatenate((B6_OndaR,silencioNota, Fsus5_OndaR,silencioNota))))

Compasso2 = np.concatenate((np.concatenate((E3_D3_OndaR,E3_D3_OndaR,E3_D3_OndaR,E3_D3_OndaR)) + np.concatenate((C5_OndaR,silencioNota, G5_OndaR,silencioNota)),
                            np.concatenate((E3_D3_OndaR,E3_D3_OndaR,E3_D3_OndaR,E3_D3_OndaR)) + np.concatenate((B6_OndaR,silencioNota, Fsus5_OndaR,silencioNota))))

Compasso5 = np.concatenate((np.concatenate((E3_Fsus3_OndaR,E3_Fsus3_OndaR,E3_Fsus3_OndaR,E3_Fsus3_OndaR)) + np.concatenate((C5_OndaR,silencioNota, G5_OndaR,silencioNota)),
                            np.concatenate((E3_Fsus3_OndaR,E3_Fsus3_OndaR,E3_Fsus3_OndaR,E3_Fsus3_OndaR)) + np.concatenate((B6_OndaR,silencioNota, Fsus5_Onda,silencioNota))))

Compasso3 = np.concatenate((C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR)) + np.concatenate(((G3_OndaR + C5_OndaR),
                                 silencioNota,(Fsus3_OndaR + G5_OndaR),silencioNota,
                                 (E3_OndaR + B6_OndaR),silencioNota,(B3_OndaR + Fsus5_OndaR),silencioNota))

Compasso6 = np.concatenate((C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR,C3_OndaR)) + np.concatenate(((B5_OndaR + C5_OndaR),
                                  silencioNota,(G3_OndaR + G5_OndaR),silencioNota,
                                 (Fsus3_OndaR + B6_OndaR),silencioNota,(B5_OndaR + Fsus5_OndaR),silencioNota))

Compasso4 = np.concatenate((np.concatenate((Csus3_Onda,Csus3_Onda,Csus3_Onda,Csus3_Onda)) + np.concatenate(((E3_OndaR + C5_OndaR),
                                  silencioNota, (Csus3_OndaR + G5_OndaR),silencioNota)),
                                   np.concatenate((Csus3_Onda,Csus3_Onda,Csus3_Onda,Csus3_Onda)) + np.concatenate(((B5_OndaR + B6_OndaR),
                                  silencioNota, (E3_OndaR + Fsus5_OndaR),silencioNota))))

Compasso7 = np.concatenate((np.concatenate((Csus4_Onda,Csus4_Onda,Csus4_Onda,Csus4_Onda)) + np.concatenate(((E3_OndaR + C5_OndaR),
                                  silencioNota, (Csus3_OndaR + G5_OndaR),silencioNota)),
                      np.concatenate((Csus4_Onda,Csus4_Onda,Csus4_Onda,Csus4_Onda)) + np.concatenate(((B3_OndaR + B6_OndaR),
                                  silencioNota, (E3_OndaR + Fsus5_OndaR),silencioNota))))

#-----------CONCATENAR AS ONDAS DE ACORDO COM OS COMPASSOS-----------
onda_concatenada = np.concatenate(( Compasso0,
#----------------------------------------------------------------
                                    Compasso1,
                                    Compasso2,
                                    Compasso3,
                                    Compasso4,
#----------------------------------------------------------------
                                    Compasso1,
                                    Compasso5,
                                    Compasso6,
                                    Compasso7,
#----------------------------------------------------------------
                                    Compasso1,
                                    Compasso2,
                                    Compasso3,
                                    Compasso4,
#----------------------------------------------------------------
                                    Compasso0
))



# Filtra o rudo
onda_filtrada = sp.wiener(onda_concatenada, mysize=15)

nome_arquivo = 'LavenderTown.wav'
wavfile.write(nome_arquivo, 44100, onda_filtrada.astype(np.float32))

# Ler o arquivo de udio original para filtragem
audio_file = 'LavenderTown.wav'
audio, samplerate = sf.read(audio_file)

# Definir as configuraes do filtro passa-baixa
frequencia_corte = 400  # Frequncia de corte em Hz
ordem = 10  # Ordem do filtro

# Projetar o filtro passa-baixa
b, a = butter(ordem, frequencia_corte / (samplerate / 2), btype='low')

# Aplicar o filtro passa-baixa ao udio
audio_filtrado = filtfilt(b, a, audio)

# Salvar o udio filtrado em um novo arquivo
audio_filtrado_file = 'LavenderTownFiltrado.wav'
sf.write(audio_filtrado_file, audio_filtrado, samplerate)

# Plote os sinais
plt.figure(figsize=(50, 10))
plt.plot(tempo_ajustado, onda_filtrada)
plt.title('Lavender town')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()



```

```python
# Espectogramas

# Gerar e exibir o espectrograma do udio original
audio_file = 'LavenderTown.wav'
y, sr = librosa.load(audio_file)

plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
librosa.display.specshow(D, y_axis='log', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma de Lavender Town')
plt.show()

# Gerar e exibir o espectrograma do udio filtrado
audio_file = 'LavenderTownFiltrado.wav'
y, sr = librosa.load(audio_file)

plt.figure(figsize=(10, 4))
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
librosa.display.specshow(D, y_axis='log', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma de Lavender Town Filtrado')
plt.show()
```
